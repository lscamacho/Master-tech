```python
import pandas as pd # pandas para estruturação
import matplotlib.pyplot as plt # gráfico
import seaborn as sns # gráfico
import atair as alt
```

```python
df_dados = pd.read_csv('gastos-deputados.csv', sep=';')
df_dados = df_dados.drop_duplicates(subset ='txNomeParlamentar')
df_dados.head()
```

```python
qtd_candidatos_uf = df_dados.groupby('sgUF', as_index=False).count()[['ideCadastro','sgUF']]
qtd_candidatos_uf
```

```python
import altair as alt
```

```python
alt.Chart(qtd_candidatos_uf).mark_bar().encode(
    x= alt.X('sgUF', title='Estado'),
    y= alt.Y('ideCadastro', title='Qtd de Políticos'),
    color=alt.Color('sgUF', legend=None)
).interactive()
```

```python
qtd_candidatos_part = df_dados.groupby('sgPartido', as_index=False).count()[['ideCadastro', 'sgPartido']]
qtd_candidatos_part
```

```python
alt.Chart(qtd_candidatos_part, title ='QTD CANDIDATOS POR PARTIDO').mark_bar().encode(
    x = alt.X('sgPartido', title='Partido'),
    y = alt.Y('ideCadastro', title='Quantidade'),
    color = alt.Color('ideCadastro', legend=None)
).interactive()
```
